```python
import jax.numpy as jnp
import chex
import numpy as np
from os import PathLike
import pandas as pd
```

```python
chex.assert_gpu_available()
```

```python
def load_data(path: PathLike):
    df = pd.read_csv(path, sep="\t", names=["X", "Y", "Wave", "Intensity"], comment="#")
    repetitions = df["Wave"].value_counts()[df["Wave"].iloc[0]]
    raman_shift = np.asarray(np.split(df["Wave"].values, repetitions))
    row_perm = np.argsort(raman_shift, axis = 1)
    raman_shift = raman_shift[:, row_perm]
    counts = np.asarray(np.split(df["Intensity"].values, repetitions))
    sorted_counts = []
    for row_idx, row in enumerate(counts):
        sorted_counts.append(row[row_perm[row_idx]])
    counts = np.stack(sorted_counts)
    raman_shift = raman_shift
    return {"raman_shift": raman_shift, "counts": counts}
```

```python
dataset = load_data('../data/Mixed_8x8.txt')
```
