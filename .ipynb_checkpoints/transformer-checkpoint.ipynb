```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import torch 
```

```python
def remove_cosmic_ray_peaks(y):
    spectrum = y.copy()
    ker=3
    l=len(spectrum)
    
    for i in range(l):
        if i>=ker and i<l-ker:
            m=np.mean(spectrum[i-ker:i+ker])
            if spectrum[i] > m*1.5:
                spectrum[i] = (spectrum[i-ker]+spectrum[i+ker])/2
            
                
    return spectrum
```

```python
df = pd.read_csv("map_c_100x10.txt", sep='\t', names=["X", "Y", "Wave", "Intensity"], comment='#')
repetitions = df['Wave'].value_counts()[df['Wave'].iloc[0]]
r_shift = np.asarray(np.split(df['Wave'].values, repetitions))
spectra = np.asarray(np.split(df['Intensity'].values, repetitions))
```

```python
x=r_shift[1,:]
y=spectra[1,:]
```

```python
r_shift=torch.tensor(r_shift)
spectra = torch.tensor(spectra)
```

```python
condition = torch.logical_not(torch.logical_and(r_shift>1525, r_shift < 1650))
filt=torch.where(condition, spectra, -1)
filtered_data = torch.stack([filt, r_shift], dim=-1).float()
true_data = torch.stack([spectra, r_shift], dim=-1).float()
```

```python
encoder = torch.nn.TransformerEncoderLayer(2,2,dim_feedforward=64)
transformer = torch.nn.TransformerEncoder(encoder,2)
```

```python
out = transformer(filtered_data[:8])
```
