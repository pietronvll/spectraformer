```python
import jax
```

```python
key = jax.random.key(seed=1)

prngkey = jax.random.PRNGKey(seed=1)

print(key)
print(key.shape)
print(type(key))

print(prngkey)
print(prngkey.shape)
print(type(prngkey))
```

```python
dropout_key = jax.random.key(seed=1)
rng_streams = {"dropout": dropout_key}
num_devices = 1
dropout_sharded = jax.random.split(rng_streams["dropout"], num_devices)
print(dropout_sharded)
```
